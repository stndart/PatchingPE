```python
from typing import cast

import lief
```

```python
pe_lief = cast(lief.PE.Binary, lief.PE.parse("ASLRtest.exe"))
```

```python
rva = 0x405060 - pe_lief.optional_header.imagebase
block_base = rva & (~0xfff)
hex(block_base), hex(rva - block_base)
```

```python
reloc = lief.PE.Relocation()
reloc.virtual_address = block_base

reloc_entry = lief.PE.RelocationEntry()
reloc_entry.type = lief.PE.RelocationEntry.BASE_TYPES.HIGHLOW
reloc_entry.position = rva - block_base

reloc.add_entry(reloc_entry)
pe_lief.add_relocation(reloc)
```

```python
config = lief.PE.Builder.config_t()
config.relocations = True

bb = lief.PE.Builder(pe_lief, config)
bb.build()
bb.write("ASLRtest2.exe")
```
